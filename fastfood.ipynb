```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Nutrition_Value_Dataset.csv")
```

```python
df.tail()
```

```python
df.describe()
```

```python
df.isnull().sum
```

```python
df['Company'].unique()
df['Category'].unique()
# df['Product'].unique() # This might be too long for many products
```

```python
df['Company'].value_counts()
df['Category'].value_counts()
```

```python
df['Energy (kCal)'].fillna(df['Energy (kCal)'].mean(), inplace=False)
```

```python
 df['Energy (kCal)'] = pd.to_numeric(df['Energy (kCal)'], errors='coerce')
```

```python
df.rename(columns={'Per Serve Size': 'Per_Serve_Size_g'}, inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Energy (kCal)'], kde=True, bins=30)
plt.title('Distribution of Energy (kCal)')
plt.xlabel('Energy (kCal)')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_nutrition_by_company = df.groupby('Company')[['Energy (kCal)', 'Carbohydrates (g)', 'Protein (g)', 'Total Fat (g)', 'Sugar (g)']].mean().sort_values(by='Energy (kCal)', ascending=False)

plt.figure(figsize=(12, 7))
sns.barplot(x=avg_nutrition_by_company.index, y=avg_nutrition_by_company['Energy (kCal)'])
plt.title('Average Energy (kCal) by Company')
plt.xlabel('Company')
plt.ylabel('Average Energy (kCal)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
avg_nutrition_by_category = df.groupby('Category')[['Energy (kCal)', 'Carbohydrates (g)', 'Protein (g)', 'Total Fat (g)', 'Sugar (g)']].mean().sort_values(by='Energy (kCal)', ascending=False)

plt.figure(figsize=(12, 7))
sns.barplot(x=avg_nutrition_by_category.index, y=avg_nutrition_by_category['Protein (g)'])
plt.title('Average Protein (g) by Category')
plt.xlabel('Category')
plt.ylabel('Average Protein (g)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Select only numerical columns for correlation
numerical_cols = ['Energy (kCal)', 'Carbohydrates (g)', 'Protein (g)', 'Fiber (g)',
                  'Sugar (g)', 'Total Fat (g)', 'Saturated Fat (g)', 'Trans Fat (g)',
                  'Cholesterol (mg)', 'Sodium (mg)']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Nutritional Values')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total Fat (g)', y='Energy (kCal)', data=df, hue='Company', alpha=0.7)
plt.title('Energy (kCal) vs. Total Fat (g)')
plt.xlabel('Total Fat (g)')
plt.ylabel('Energy (kCal)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
pizza_hut_data = df[df['Company'] == 'Pizza Hut']
dominos_data = df[df['Company'] == 'Dominos']

print("\nPizza Hut Average Nutrition:")
print(pizza_hut_data[numerical_cols].mean())

print("\nDominos Average Nutrition:")
print(dominos_data[numerical_cols].mean())
```

```python
dessert_data = df[df['Category'] == 'Dessert']
print("\nDessert Nutrition Statistics:")
print(dessert_data[numerical_cols].describe())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Per_Serve_Size_g', y='Energy (kCal)', data=df)
plt.title('Energy (kCal) vs. Per_Serve_Size_g')
plt.xlabel('Per_Serve_Size_g')
plt.ylabel('Energy (kCal)')
plt.grid(True)
plt.show()
```

```python
company_counts = df['Company'].value_counts()

plt.figure(figsize=(7, 8))
plt.pie(company_counts, labels=company_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Products by Company')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.savefig('company_pie_chart.png')
plt.show()
```

```python
category_counts = df['Category'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Products by Category')
plt.axis('equal')
plt.tight_layout()
plt.savefig('category_pie_chart.png')
plt.show()
```

```python
# List of numerical columns for analysis
numerical_cols = ['Energy (kCal)', 'Carbohydrates (g)', 'Protein (g)', 'Sugar (g)', 'Total Fat (g)', 'Sodium (mg)']

for col in numerical_cols:
    plt.figure(figsize=(12, 7))
    sns.boxplot(x='Category', y=col, data=df)
    plt.title(f'{col} Distribution by Category')
    plt.xlabel('Category')
    plt.ylabel(col)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
print("\n--- Nutritional Value Distribution by Company (Violin Plots) ---")

for col in numerical_cols:
    plt.figure(figsize=(12, 7))
    sns.violinplot(x='Company', y=col, data=df)
    plt.title(f'{col} Distribution by Company')
    plt.xlabel('Company')
    plt.ylabel(col)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
print("\n--- Scatter Plot: Energy (kCal) vs. Sugar (g) colored by Category ---")
plt.figure(figsize=(12, 8))
sns.lmplot(x='Sugar (g)', y='Energy (kCal)', hue='Category', data=df, aspect=1.5, ci=None, scatter_kws={'alpha':0.6})
plt.title('Energy (kCal) vs. Sugar (g) by Category (with Regression Line)')
plt.xlabel('Sugar (g)')
plt.ylabel('Energy (kCal)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
